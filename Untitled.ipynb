```python
my_str = "Nishant is a great mentor. nishant is down to earth"

count_dict = {}

for word in my_str.split():
    word_lower = word.lower()
    if word_lower in count_dict:
        count_dict[word_lower]+=1
    else:
        count_dict[word_lower] = 1
print(count_dict)
```

```python
my_dict = {"a":1,"b":1,"c":3}
reversed_dict = {}
is_reverse = True
for key, value in my_dict.items():
    if value in reversed_dict.keys():
        is_reverse = False
        break
    else:
        reversed_dict[value] = key

if is_reverse:
    print(reversed_dict)
else:
    print("Dicitonary can't be reversed as duplicate values are present")
```

```python
for n in range(2, 101):
    is_prime = True
    for i in range(2, n):
        if n % i == 0:
            is_prime = False
            break
    if is_prime:
        print(n)
```
